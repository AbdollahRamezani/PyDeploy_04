```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
np_array = np.frombuffer(response.content)  #برای اینکه عکس بتوان دیکد کرد ابتدا باید به نامپای ارری تبدیل شود
image = cv2.imdecode(np_array, cv2.IMREAD_UNCHANGED)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.show()
```
